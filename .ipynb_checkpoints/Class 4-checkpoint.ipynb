```python
class Node:
    def __init__(self, data):
        self.next = None
        self.prev = None
        self.data = data
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def append(self, value):
        new_node = Node(value)
        
        if not self.head:
            self.head = new_node
            self.tail = self.head
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            self.tail = new_node
            new_node.prev = current
    
    # Add same methods used for SinglyLinkedList in class 3
    
    # insert
    def insert(self, index, value):
        new_node = Node(value)
        
        if not self.head:
            self.head = new_node
            self.tail = self.head
        elif index == 0:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        else:
            count = 0
            current = self.head
            while current.next and count != index:
                count += 1
                current = current.next
            if count == index and count != (len(self) - 1):
                current.prev.next = new_node
                new_node.prev = current.prev
                current.prev = new_node
                new_node.next = current
            elif not current.next:
                current.next = new_node
                new_node.prev = current
                self.tail = new_node
    
    # remove
    def remove(self, value):
        if not self.head:
            raise ValueError(f"The list is emtpy. {value} does not exist in the list.")
        if self.head.data == value:
            if not self.head.next:
                self.head == None
                self.tail == None
            else:
                self.head = self.head.next
                self.head.prev = None
        else:
            current = self.head
            while current.next and current.data != value:
                current = current.next
            if not current.next and current.data != value:
                raise ValueError(f"{value} does not exist in the list.")
            elif not current.next:
                self.tail = current.prev
                self.tail.next = None
                current.prev = None
            else:
                current.prev.next = current.next
                current.next.prev = current.prev
                current.prev = None
                current.next = None
            
    
    # __len__
    def __len__(self):
        count = 0
        if not self.head:
            return count
        else:
            current = self.head
            while current:
                count += 1
                current = current.next
            return count
    
    # count
    def count(self, value):
        if not self.head:
            return 0
        else:
            count = 0
            current = self.head
            while current:
                if current.data == value:
                    count += 1
                current = current.next
            return count
    
    # index
    def index(self, value):
        if not self.head:
            raise ValueError(f"The list is empty. {value} does not exist in the list.")
        count = 0
        current = self.head
        while current:
            if current.data == value:
                return count
            count += 1
            current = current.next
        else:
            raise ValueError(f"{value} not found in the list.")
    
    def __str__(self):
        out = "[%s" % self.head.data
        current = self.head.next
        while current:
            out += ", %s" % current.data
            current = current.next
        out += "]"
        return out
            
```

```python
memo = {}

def fib_memo(value):
    if value in memo:
        return memo[value]
    if value == 1:
            value = 1
    elif value == 2:
            value = 1
    elif value > 2:           
            new_value =  fib_memo(value -1) + fib_memo(value -2)
    memo[value] = new_value
    return value
```
